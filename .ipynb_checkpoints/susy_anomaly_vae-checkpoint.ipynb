```python
class Gauss(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a digit."""

    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        #return z_mean + tf.exp(0.5 * z_log_var) * epsilon
        return z_mean + tf.exp(-1*(z_log_var)*(z_log_var)) * epsilon

class Poisson(layers.Layer):
    def call(self, inputs):
        z_mean = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        #return z_mean + tf.exp(0.5 * z_log_var) * epsilon
        return z_mean + tf.exp(-1*z_mean) * epsilon + tf.exp(-1*z_mean*z_mean) * epsilon
```

```python
i = 0
latent_dim = 5
df = df_low_level

df_train = df[df['label']==0]
df_test = df[df['label']==1]
X = df_train.drop('label',axis=1)

X_train, X_valid =  train_test_split(X,random_state=1,test_size=.3)
X_valid, X_test =  train_test_split(X_valid,random_state=1,test_size=.4)
input_shape = X_train.shape[1]
#scaler = StandardScaler()
#scaler.fit_transform(X_train)
#scaler.transform(X_valid)
#scaler.transform(X_test)
inputs = Input(shape=(input_shape))
x = Dense(75, activation='LeakyReLU',use_bias=False)(inputs)
x = Dense(50, activation='LeakyReLU',use_bias=False)(x)
x = Dense(25, activation='LeakyReLU',use_bias=False)(x)

#z_mean_1 = layers.Dense(latent_dim,activation='LeakyReLU', name="z_mean_1",use_bias=False)(x)
z_mean_2 = layers.Dense(latent_dim,activation='LeakyReLU', name="z_mean_2",use_bias=False)(x)
#z_log_var = layers.Dense(latent_dim, name="z_log_var",use_bias=False)(x)
#z_1 = Gauss()([z_mean_1, z_log_var])
z_2 = Poisson()(z_mean_2)

#merged = Concatenate()([z_1, z_2])
#x = Activation("LeakyReLU")(merged)
bottle_neck = Dense(latent_dim, activation='LeakyReLU',use_bias=False)(z_2)

x = Dense(20, activation='LeakyReLU',use_bias=False)(bottle_neck)
x = Dense(50, activation='LeakyReLU',use_bias=False)(x)
x = Dense(75, activation='LeakyReLU',use_bias=False)(x)
decoded = Dense(input_shape)(x)

susy_coder = Model(inputs=inputs, outputs=decoded)
susy_coder.summary()
tf.keras.utils.plot_model(
    susy_coder,
    to_file="model.png")

susy_coder.compile(optimizer=keras.optimizers.Adam(learning_rate=0.003),loss="mse",metrics=['ACC'])
model_name = "AutoEncoder"
susy_coder.fit(X_train,X_train,epochs=30,batch_size=512,validation_data=(X_valid, X_valid))

#z_mean, z_log_var, z = encoder(data)

"""
x = Dense(200, activation='ReLU',use_bias=False)(inputs)
x = Dense(100, activation='ReLU',use_bias=False)(x)
x = Dense(50, activation='ReLU',use_bias=False)(x)
x = Dense(30, activation='ReLU',use_bias=False)(x)
encoder = Dense(6, activation='ReLU',use_bias=False)(x)
x = Dense(30, activation='ReLU',use_bias=False)(encoder)
x = Dense(50, activation='ReLU',use_bias=False)(x)
x = Dense(100, activation='ReLU',use_bias=False)(x)
x = Dense(200, activation='ReLU',use_bias=False)(x)


x = Dense(20, activation='LeakyReLU',use_bias=False)(inputs)
x = Dense(15, activation='LeakyReLU',use_bias=False)(x)
encoder = Dense(4, activation='LeakyReLU',use_bias=False)(x)
x = Dense(15, activation='LeakyReLU',use_bias=False)(encoder)
x = Dense(20, activation='LeakyReLU',use_bias=False)(x)

#decoded = Dense(input_shape)(x)
susy_coder = Model(inputs=inputs, outputs=decoded)
susy_coder.summary()
susy_coder.compile(optimizer=keras.optimizers.Adam(learning_rate=0.002),loss="mse",metrics=['ACC'])
model_name = "AutoEncoder"
susy_coder.fit(X_train,X_train,epochs=10,batch_size=512,validation_data=(X_valid, X_valid))
#save = pd.DataFrame({"y_pred" : y_pred.ravel(),"y_true" : y_valid.ravel()})
#save.to_csv("data_"+model_name+"_"+levels[i]+".csv", index=False)
#    i = i + 1
"""
```

```python
def mse_loss(true, prediction):
    loss = tf.reduce_mean(tf.math.square(true - prediction), axis=-1)
    return loss

total_loss = []
total_loss.append(mse_loss(X_test, bkg_prediction.astype(np.float32)).numpy())
total_loss.append(mse_loss(signal_data, signal_results[i][2].astype(np.float32)).numpy())

bin_size = 100


target_background = np.zeros(total_loss[0].shape[0])

plt.figure(figsize=(10, 8))

trueVal = np.concatenate((np.ones(total_loss[1].shape[0]), target_background))  # anomaly=1, bkg=0
predVal_loss = np.concatenate((total_loss[1], total_loss[0]))

fpr_loss, tpr_loss, threshold_loss = roc_curve(trueVal, predVal_loss)

auc_loss = auc(fpr_loss, tpr_loss)

plt.plot(
    tpr_loss,
    1-fpr_loss,
    "-",
    label=f"{label} (auc = {auc_loss * 100.0:.1f}%)",
    linewidth=1.5,
)

#plt.semilogx()
#plt.semilogy()
plt.ylabel("Background Rejection")
plt.xlabel("Signal Efficiency")
plt.legend(loc="center right")
#plt.grid(True)
#plt.tight_layout()
plt.plot(np.linspace(0, 1), np.linspace(1, 0), "--", color="0.75")
#plt.axvline(
#0.00001, color="red", linestyle="dashed", linewidth=1
#)  # threshold value for measuring anomaly detection efficiency
plt.title("ROC AE")
#plt.show()
```
